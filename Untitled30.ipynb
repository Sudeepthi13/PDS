#2)Handling with missing values

We will first display all the missing values

```python
# Display missing values before replacement
print("Missing values before replacement:")
print(df.isna().sum())
```

```python
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
```

The missing values are replaced with the calculated mean for each respective numeric column.

```python
for column in numeric_columns:
  if df[column].isna().sum() > 0:
    mean_value = df[column].mean()
    df[column].fillna(mean_value, inplace=True)
```

```python
print("\nMissing values after replacement:")
print(df.isna().sum())
```

Here now we can see the missing values in numeric columns (bilirubin, alk_phosphate, sgot, albumin, protime) are filled with mean values. After replacement there are 0 missing values.
